```python
import gradio as gr
from PIL import Image
import torch
from transformers import BlipProcessor, BlipForConditionalGeneration
from diffusers import StableDiffusionPipeline
import traceback

# Load models (fastest: BLIP + SD-Turbo)
device = "cuda" if torch.cuda.is_available() else "cpu"

blip_processor = BlipProcessor.from_pretrained("Salesforce/blip-image-captioning-base")
blip_model = BlipForConditionalGeneration.from_pretrained(
    "Salesforce/blip-image-captioning-base"
).to(device)

pipe = StableDiffusionPipeline.from_pretrained(
    "stabilityai/stable-diffusion-turbo",
    torch_dtype=torch.float16 if torch.cuda.is_available() else torch.float32
).to(device)
pipe.safety_checker = None  # safety filter off (demo purposes)

def caricature_from_photo(image):
    try:
        raw = image.convert('RGB')
        inputs = blip_processor(raw, return_tensors="pt").to(device)
        out = blip_model.generate(**inputs, max_new_tokens=30)
        desc = blip_processor.decode(out[0], skip_special_tokens=True)

        # Caricature-style prompt
        prompt = (
            f"{desc}, cute cartoon, gentle lines, charming and lovely, "
            "sweet expression, digital illustration, simple background, cheerful 😄✨"
        )

        img = pipe(
            prompt,
            num_inference_steps=4,
            guidance_scale=0.0
        ).images[0]

        info = (
            f"**Image description**: {desc}\n\n"
            f"**Prompt used**: `{prompt}`"
        )
        return img, info

    except Exception as e:
        err = f"❗️ Error!\n{e}\n\n{traceback.format_exc()}"
        blank = Image.new("RGB", (512,512), "#FFFDE7")
        return blank, err

custom_css = """
body { background: linear-gradient(135deg, #a7ffeb 0%, #fce4ec 100%); }
.gradio-container { background-color: #fafcff !important; }
.gr-button { background: #ffd600; color: #252525; border-radius:18px; font-size:1.2em; }
.gr-button:hover { background:#ffd54f; }
#preview img {
    border-radius: 30px; border: 4px solid #ffff8d;
    box-shadow: 0px 8px 56px #ffd60044; margin-top: 1em;
}
#main .block {
    background: #ffffffb8; border-radius: 22px;
    box-shadow: 0 4px 32px #b3afec33; margin:18px;
}
"""

with gr.Blocks(css=custom_css, theme=gr.themes.Soft()) as demo:
    gr.Markdown("""
    <div style="text-align:center; padding-top:.5em;">
      <span style="font-size:2.3em;">🎨 <b style="color:#1ec69d">Caricature Magic</b> App ✨</span>
      <p style="font-size:1.11em; color:#575d76;">
        Turn any photo into a caricature in 1 second!<br>
        Upload from <b>Camera 📷</b> or <b>Album</b>.
      </p>
    </div>
    """)
    with gr.Row(elem_id="main"):
        with gr.Column():
            photo = gr.Image(label="📷 Capture or Upload Photo", type="pil", height=300)
            btn = gr.Button("👉 Create Caricature!")
        with gr.Column():
            preview = gr.Image(label="🖼️ Preview", elem_id="preview", height=350)
            info = gr.Markdown(label="Details", value="")
    btn.click(caricature_from_photo, inputs=photo, outputs=[preview, info])

    gr.Markdown("""
    <div style="background:#fffde7; border-radius:18px; border:1.5px dashed #ffe082; padding:1em; text-align:center; margin-top:18px;">
      <b>💡 How to Use</b><br/>
      <ul style="text-align:left; margin-left:1.8em;">
        <li>On mobile, choose Camera or Album.</li>
        <li>Click “Create Caricature!” for instant results.</li>
        <li>Long-press or right-click result to save/share/print.</li>
      </ul>
    </div>
    """)

demo.launch()
```
